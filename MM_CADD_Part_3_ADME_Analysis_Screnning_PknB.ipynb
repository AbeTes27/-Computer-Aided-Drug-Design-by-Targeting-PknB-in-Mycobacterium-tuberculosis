## Molecular Modeling Project *Serine/threonine-protein kinase PknB in Mycobacterium tuberculosis*

```python
import pandas as pd
```

```python
df = pd.read_csv('swissadme.csv')
```

```python
df.head(50)
df.columns
```

```python
df = df.loc[(df['MW'] < 500)]
```

```python
df = df.loc[((df['XLOGP3'] > -0.4) & (df['XLOGP3'] < 5.0))]
```

```python
df = df.loc[((df['#Rotatable bonds'] >= 0) & (df['#Rotatable bonds'] <= 9))]
```

```python
df = df.loc[(df['#H-bond acceptors'] < 10)]
```

```python
df = df.loc[(df['#H-bond donors'] < 5)]
```

```python
df = df.loc[(df['TPSA'] < 120)]
```

```python
df = df.loc[((df['ESOL Log S'] > -6) & (df['ESOL Log S'] < 0))]
```

```python
df = df.loc[((df['Fraction Csp3'] >= 0.25) & (df['Fraction Csp3'] <= 1))]
```

```python
df.to_csv('ADME_clean_final.csv', index=False)
df
```
